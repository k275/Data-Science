Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
Ans->A scatter plot is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data. If the points are coded, one additional variable can be displayed

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('scatter Plot')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
Ans-we use subplot() function to draw multiple plots in one figure.

```python
import matplotlib.pyplot as plt
import numpy as np
##line-1
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(4, 1, 1)
plt.plot(x,y)
plt.title("First plot")

##line-2
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(4, 1, 2)
plt.plot(x,y)
plt.title("Second plot")

##line-3
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(4, 1, 3)
plt.plot(x,y)
plt.title("Third plot")

##line-4
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(4, 1, 4)
plt.plot(x,y)
plt.title("Fourth plot")
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
Ans-The purpose of a bar graph is to convey relational information quickly in a visual manner. The bars display the value for a particular category of data. The vertical axis on the left or right side of the bar graph is called the y-axis. The horizontal axis at the bottom of a bar graph is called the x-axis.

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel('Company name')
plt.ylabel('Profit')
plt.title("Companies profits Chart")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
Ans-
A box plot is a graphical representation of a data set that displays the median, quartiles, and outliers of the data. It is also known as a box-and-whisker plot because it is composed of a box that represents the interquartile range (IQR) of the data, and whiskers that extend from the box to show the range of the data outside the IQR.

The box in a box plot represents the middle 50% of the data, with the lower boundary of the box being the first quartile (Q1) and the upper boundary being the third quartile (Q3). The line inside the box represents the median. The whiskers extend from the box to the minimum and maximum values within 1.5 times the IQR from the edges of the box. Any data points outside the whiskers are considered outliers.

Box plots are used to summarize and compare the distribution of numerical data. They are particularly useful when comparing multiple datasets, as they provide a clear visual representation of differences in the central tendency, variability, and skewness of the data. Box plots can also reveal the presence of outliers, which can affect the interpretation of summary statistics like the mean and standard deviation.

Overall, box plots are a valuable tool for exploratory data analysis and for communicating the characteristics of a data set to others.

```python
import matplotlib.pyplot as plt
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2])
plt.xlabel('box')
plt.ylabel('value')
plt.title('box-plot')
plt.show()
```
